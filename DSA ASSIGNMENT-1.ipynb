#### LINKED LIST :

* It is a LINEAR datastructure in which data can be store in non-consequent manner such that the elements are connected through the pointers i.e Chain of Nodes. NODE contains a value and pointer to the next Node.

* Since PYTHON library doesn't have LINKED LIST they are implemented with help of nodes.

```python
# A single node of a singly linked list
class Node:
  # constructor
  def __init__(self, data = None, next=None): 
    self.data = data
    self.next = next

# A Linked List class with a single head node
class LinkedList:
  def __init__(self):  
    self.head = None
  
  # insertion method for the linked list
  def insert(self, data):
    newNode = Node(data)
    if(self.head):
      current = self.head
      while(current.next):
        current = current.next
      current.next = newNode
    else:
      self.head = newNode
  
  # print method for the linked list
  def printLL(self):
    current = self.head
    while(current):
      print(current.data)
      current = current.next

# Singly Linked List with insertion and print methods
LL = LinkedList()
LL.insert(3)
LL.insert(4)
LL.insert(5)
LL.printLL()
```

#### ALGORITHM :

* **ALGORITHM** can be defined as the set of commamnds/set of insturctions which allows the system to process the given input and return thedesired output upon execution of the written instructions/commands.

* ALGORITHM provides the needed mechanism for processing the problem and to define the logic required for processing the problem

* The stages innvolved in writting an ALGORITHM are
                        INPUT   PROCESSING COMMANDS   OUTPUT

* It is an **ALGORITHM** to find **PYTHAGOREAN TRIPLETS**

```python
## INPUT
Naturalnumber=int(input("Enter upper limit:"))

## PROCESSING COMMANDS
c=0
m=2
while(c<Naturalnumber):
    for n in range(1,m+1):
        a=m*m-n*n
        b=2*m*n
        c=m*m+n*n
        if(c>Naturalnumber):
            break
        if(a==0 or b==0 or c==0):
            break
## OUTPUT
        print(a,b,c)
    m=m+1 ## ALSO PART OF PROCESSING AFTER OUTPUT
```

# 5. Describe the Tree traversal method?


#### TREE TRAVERSAL METHOD

* TREE TRAVERSAL method is a HIERARCHICAL DATA STRUCTURE. Unlike ARRAYS,STACKS AND QUEUE,LINKED LISTS which have only one way to traverse inorder to search an element. But in Tree Traversal we visit all nodes and may print their values in different ways.
* TREE being a non-linear data structure which allows quicker and easier access to data  we search. Inorder to perform certain operations on Tree data structure we require TREE TRAVERSAL ALGORITHM to achieve our objective.

* TREE TRAVERSAL ALGORITHM can be broadly classified into two types by the order of which they visit the nodes.
            DEPTH FIRST SEARCH(DFS)     BREADTH FIRST SEARCH(BFS)

#### DEPTH FIRST SEARCH(DFS)

* In this searching ALGORITHM, it starts with root node first and and visit all the nodes of one branch as deep as possible of the choosen node and before backtracking, it visits all other branches in a similar fashion. There are three subtypes under this
1. Pre-order Traversal (Root, Left ,
Right)
2. In-order Traversal (Left ,Root,
Right).
3. Post-order Traversal (Left , Right,
Root)

#### BREADTH FIRST SEARCH(BFS) :


* In this searching ALGORITHM, It also starts from the root node and visits all nodes of current depth before moving to the next depth in the tree.
        Level-Order Traversal

#### Pre-Order Traversal :

* In this Tree Traversal Method, Each node of the tree is traversed before its Left and Right sub-trees are traversed.This is also called as "Back-Tracking".
* Steps of traversing a Tree by Pre-Order Traversal method:
  1. Visit the root node.
  2. Visit the left tree.
  3. Visit the right tree

#### In-Order Traversal :

* In this Tree Traversal Method, Each  root node of the  sub-tree is traversed after its Left tree traversed and but before its Right sub-tree is  traversed.
* Steps of traversing a Tree by In-Order Traversal method:
  1. Visit the left tree.
  2. Visit the root node.
  3. Visit the right tree

#### Post-Order Traversal :

* In this Tree Traversal Method, Each root node of the  sub-tree is traversed after its Left tree and Right sub-tree is  traversed.
* Steps of traversing a Tree by In-Order Traversal method:
  1. Visit the left tree.
  2. Visit the right tree.
  3. Visit the root node.

#### Level-Order Traversal :


* In this Tree Traversal Method, Each root node of the tree is visited level-by-leveland going from left to right.It requires a Queue data structure.

#### In-Order Traversal :

* In this Tree Traversal Method, Each  root node of the  sub-tree is traversed after its Left tree traversed and but before its   Right sub-tree is  traversed.
* Steps of traversing a Tree by In-Order Traversal method:
  1. Visit the left tree.
  2. Visit the root node.
  3. Visit the right tree

#### Post-Order Traversal :

* In this Tree Traversal Method, Each root node of the  sub-tree is traversed after its Left tree and Right sub-tree is  traversed.
* Steps of traversing a Tree by In-Order Traversal method:
  1. Visit the left tree.
  2. Visit the right tree.
  3. Visit the root node.

```python
# Python program to for tree traversals

# A class that represents an individual node in a
# Binary Tree


class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key


# A function to do inorder tree traversal
def Inorder(root):

    if root:
        Inorder(root.left)
        print(root.val,end=" "),
        Inorder(root.right)


# A function to do postorder tree traversal
def Postorder(root):

    if root:
        Postorder(root.left)
        Postorder(root.right)
        print(root.val,end=" "),



# Driver code
root = Node(10)
root.left = Node(28)
root.right = Node(39)
root.left.left = Node(14)
root.left.right = Node(58)

print ("Inorder traversal of binary tree is")
Inorder(root)

print ("\nPostorder traversal of binary tree is")
Postorder(root)
```
