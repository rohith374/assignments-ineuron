```python
vote_count=netflix[netflix["vote_count"].notnull()]["vote_count"].astype(int)
vote_average=netflix[netflix["vote_average"].notnull()]["vote_average"].astype(int)
vote_average
C = vote_average.mean()
C
```

```python
netflix['genres'] = netflix['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
netflix.head()
```

```python
netflix['year'] = pd.to_datetime(netflix['release_date'], errors='coerce').apply(lambda x: str(x).split('-')[0] if x != np.nan else np.nan)
```

```python
qualified = netflix[(netflix['vote_count'] >= m) & (netflix['vote_count'].notnull()) & (netflix['vote_average'].notnull())][['title', 'year', 'vote_count', 'vote_average', 'popularity', 'genres']]
qualified['vote_count'] = qualified['vote_count'].astype('int')
qualified['vote_average'] = qualified['vote_average'].astype('int')
qualified.shape
```

```python
def weighted_rating(x):
    v = x['vote_count']+1 # added +1 - Dan
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

netflix['weighted_rating'] = netflix.apply(weighted_rating, axis=1)
```

```python
qualified['wr'] = qualified.apply(weighted_rating, axis=1)
```

```python
qualified = qualified.sort_values('wr', ascending=False).head(250)
```

```python
g = netflix.apply(lambda x: pd.Series(x['genres']),axis=1).stack().reset_index(level=1, drop=True)
g.name = 'genre'
gen_netflix = netflix.drop('genres', axis=1).join(g)
```

```python
def build_chart(genre, percentile=0.85):
    netflix= gen_netflix[gen_netflix['genre'] == genre]
    vote_counts = netflix[netflix['vote_count'].notnull()]['vote_count'].astype('int')
    vote_averages = netflix[netflix['vote_average'].notnull()]['vote_average'].astype('int')
    C = vote_averages.mean()
    m = vote_counts.quantile(percentile)
    
    qualified = netflix[(netflix['vote_count'] >= m) & (netflix['vote_count'].notnull()) & (netflix['vote_average'].notnull())][['title', 'year', 'vote_count', 'vote_average', 'popularity']]
    qualified['vote_count'] = qualified['vote_count'].astype('int')
    qualified['vote_average'] = qualified['vote_average'].astype('int')
    
    qualified['wr'] = qualified.apply(lambda x: (x['vote_count']/(x['vote_count']+m) * x['vote_average']) + (m/(m+x['vote_count']) * C), axis=1)
    qualified = qualified.sort_values('wr', ascending=False).head(250)
    
    return qualified
```
